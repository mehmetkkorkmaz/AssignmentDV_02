## 2: Plot 12 most dangerous cities

```python
df=kill_dna['city'].sort_values().value_counts(sort=True)[:12]
#df=df.sort_values(ascending=False)
#df=df.sort_index(ignore_index=False)
df
```

```python
kill.dropna()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=df.index,y=df.values, ci=True)

plt.title('Most dangerous cities',color='blue')
plt.xticks(rotation=45);
```

```python
df.duplicated()
```

```python
# with seaborn barplot

```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='city', data=kill_dna,
              order=kill_dna.city.value_counts(sort=True).iloc[:12].index)

plt.title('Most dangerous cities',color='blue')
plt.xticks(rotation=45);
```

```python
# with seaborn countplot
```
